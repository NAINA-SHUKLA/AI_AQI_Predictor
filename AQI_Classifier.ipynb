```python
# ===============================
# AQI CATEGORY PREDICTION MODEL
# ===============================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# -------------------------------
# 1. Load dataset
# -------------------------------
df = pd.read_csv("AQI.csv")
print("âœ… Dataset loaded")

# -------------------------------
# 2. Create AQI Category (CORRECT LOGIC)
# -------------------------------
def aqi_category(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

df["aqi_category"] = df["aqi_index"].apply(aqi_category)

print("âœ… AQI categories created")

# -------------------------------
# 3. Select Features & Target
# -------------------------------
features = [
    "pm2_5",
    "pm10",
    "co",
    "no2",
    "temp_c",
    "humidity",
    "windspeed_kph"
]

X = df[features]
y = df["aqi_category"]

# -------------------------------
# 4. Train-test split (time-aware)
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    shuffle=False
)

# -------------------------------
# 5. Train Classification Model
# -------------------------------
model = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
print("ðŸŒ± Model trained successfully")

# -------------------------------
# 6. Evaluate Model
# -------------------------------
y_pred = model.predict(X_test)

print("\nðŸ“Š Model Accuracy:", round(accuracy_score(y_test, y_pred), 2))
print("\nðŸ“‹ Classification Report:\n")
print(classification_report(y_test, y_pred, zero_division=0))

# -------------------------------
# 7. Sample Prediction (REALISTIC OUTPUT)
# -------------------------------
sample_input = pd.DataFrame([{
    "pm2_5": 480,
    "pm10": 510,
    "co": 1900,
    "no2": 120,
    "temp_c": 12,
    "humidity": 85,
    "windspeed_kph": 2
}])

predicted_category = model.predict(sample_input)[0]

print("\nðŸŒ Sample AQI Prediction:")
print("Predicted AQI Category â†’", predicted_category)
```

```python
import joblib

# Save the trained model
joblib.dump(model, "aqi_category_model.pkl")

print("âœ… Model saved as aqi_category_model.pkl")
```
